# Handling missing data

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:\\Users\\Admin\\Desktop\\dva lab\\ad_click_dataset.csv")
df.head()
```

```python
df['age'].fillna(df['age'].mean(),inplace=True)
df.head()
```

```python
df['gender'].fillna(df['gender'].mode(),inplace=True)
df.head()
```

```python
df['browsing_history'].fillna(method='ffill',inplace=True)
df.head()
```

```python
df['time_of_day'].fillna(method='bfill',inplace=True)
df.head()
```

```python
df['ad_position'].replace({'Top':1,'Side':2,'Bottom':3},inplace=True)
df['ad_position'].fillna(df['ad_position'].median(),inplace=True)
df['ad_position'].replace({1:'Top',2:'Side',3:'Bottom'},inplace=True)
df.head()
```

```python
df=df.drop(['device_type',],axis=1)
df.head()
```

# min-max and zscore

```python
(df['age']-min(df['age']))/(max(df['age'])-min(df['age']))
```

```python
s = sum(df['age'])/len(df)
v = 0
for i in df['age']:
    v += (i-s)**2
var = (v)/len(df)
print(var)
abs((df['age']-s)/var)
```
