# Visual Analysis of Impulation Techniques

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ad_click_dataset.csv').iloc[:100,:-2]
```

```python
print(len(df))
df.head()
```

```python
df['device_type'].replace({'Desktop':0,'Mobile':1,'Tablet':2},inplace = True)
df['gender'].replace({'Male':0,'Female':1,'Non-Binary':2},inplace = True)
df['ad_position'].replace({'Top':0,'Side':1,'Bottom':2},inplace = True)
df.head()
```

```python
sns.pairplot(df)
```

```python
sns.boxplot(df.iloc[:,3:])
```

```python
df['age'] = df['age'].fillna(df['age'].mean())
df['gender'] = df['gender'].fillna(df['gender'].mean())
df['device_type'] = df['device_type'].fillna(df['device_type'].mean())
df['ad_position'] = df['ad_position'].fillna(df['ad_position'].mean())
df['browsing_history'] = df['browsing_history'].fillna(method = 'bfill')
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
sns.boxplot(df.iloc[:,3:])
```

```python
df['age'] = df['age'].fillna(df['age'].median())
df['gender'] = df['gender'].fillna(df['gender'].median())
df['device_type'] = df['device_type'].fillna(df['device_type'].median())
df['ad_position'] = df['ad_position'].fillna(df['ad_position'].median())
df['browsing_history'] = df['browsing_history'].fillna(method = 'bfill')
```

```python
sns.pairplot(df)
```

```python
sns.boxplot(df.iloc[:,3:])
```

```python
df['age'] = df['age'].fillna(df['age'].mode()[0])
df['gender'] = df['gender'].fillna(df['gender'].mode()[0])
df['device_type'] = df['device_type'].fillna(df['device_type'].mode()[0])
df['ad_position'] = df['ad_position'].fillna(df['ad_position'].mode()[0])
df['browsing_history'] = df['browsing_history'].fillna(method = 'bfill')
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
sns.boxplot(df.iloc[:,3:])
```

#Data Transformation Effect

```python
print(max(df['age']))
print(min(df['age']))
df['age'] = (df['age']-min(df['age']))/(max(df['age'])-min(df['age']))
plt.scatter(df['age'],df['id'])
```

```python
mean_age = sum(df['age'])/len(df)
variance_age = sum((df['age'] - mean_age)**2)/len(df)
print(variance_age)
df['age'] = abs((df['age'] - mean_age)/variance_age)
plt.scatter(df['age'],df['id'])
```
